```python
import numpy as np
data = puzzle_in
puzzles = []
cur_puzzle = []
for line in data:
    if not line:
        puzzles.append(cur_puzzle)
        cur_puzzle = []
        continue
    cur_puzzle.append([c for c in line])
if cur_puzzle: puzzles.append(cur_puzzle)


def checkMirroring(puzzle):
    for i in range(len(puzzle)-1):
        if not(np.array_equiv(puzzle[i], puzzle[i+1])): continue
        for j in range(1,i+10):
            if (i-j < 0) or (i+1+j) >= len(puzzle): return i+1
            # print(f"{i}, {j}: {puzzle[i-j]} to {puzzle[i+1+j]}")
            if not(np.array_equiv(puzzle[i-j], puzzle[i+1+j])): break
    return 0


total = 0
for puzzle in puzzles:
    
    hscore = checkMirroring(np.asarray(puzzle))
    vscore = checkMirroring(np.transpose(puzzle))

    total += hscore*100 + vscore
    if hscore > 0 and vscore > 0: print("DISASTER")
    if hscore == 0 and vscore == 0: print("DISASTER2")
    for p in puzzle:
        print(p)
    print(hscore, vscore, total)


print(total)
# ! 27756 is too low
# ! 33517 is too low
# ! 34000 is too high
```

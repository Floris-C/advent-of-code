```python
#Day 10 AOC 23
with open("input.txt", "r") as file:
    puzzle_in = file.read().splitlines()
print(puzzle_in[:10])

with open("example.txt", "r") as file:
    example = file.read().splitlines()
print(example[:10])
```

```python
data = example
data = [[c for c in line] for line in data]

h_growth = {}
for i, line in enumerate(data): 
    # print(line)
    h_growth[i] = h_growth.get(i-1, 0) + all([c == '.' for c in line])
print(h_growth)

v_growth = {}
for i, line in enumerate([list(x) for x in zip(*data)]): # list is data but transposed
    # print(line)
    v_growth[i] = v_growth.get(i-1, 0) + all([c == '.' for c in line])
print(v_growth)

galaxy_counter = 0
galaxy_dict = {}
for y, line in enumerate(data):
    for x, val in enumerate(line):
        if val == '.': continue
        data[y][x] = str(galaxy_counter)
        galaxy_dict[galaxy_counter] = (x + v_growth[x], y + h_growth[y])
        galaxy_counter += 1

def manhattan_dist(c1, c2):
    x1,y1 = c1
    x2,y2 = c2
    return abs(x2-x1)+abs(y2-y1)

total_dist = 0
for lk, lv in galaxy_dict.items():
    for rk, rv in galaxy_dict.items():
        if rk <= lk: continue
        dist = manhattan_dist(lv, rv)
        # print(dist)
        total_dist += dist

print(f"total dist: {total_dist}")
        

# for d in data:
#     print(d)
# print(galaxy_dict)

```

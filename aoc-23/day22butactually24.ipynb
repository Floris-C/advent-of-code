```python
def parse_line(l):
    l,r = l.split('~')
    a,b,c = l.split(',')
    x,y,z = r.split(',')
    return ({'x':int(a),'y':int(b),'z':int(c)},{'x':int(x),'y':int(y),'z':int(z)})

def new_layer():
    return [['.' for i in range(10)] for j in range(10)]

def checkLayer(block, layer):
    for x in range(block[0]['x'], block[1]['x']+1):
        for y in range(block[0]['y'], block[1]['y']+1):
            if layer[y][x] != '.': return False
    return True

def place_block(block, blockI, tower):
    starting_layer = block[0]['z'] 
    block_height = block[1]['z'] - block[0]['z'] 
    isClear = (starting_layer >= 1) and (starting_layer >= len(tower) or checkLayer(block, tower[starting_layer]))
    while isClear:
        starting_layer -= 1
        isClear = (starting_layer >= 1) and (starting_layer >= len(tower) or checkLayer(block, tower[starting_layer]))
    starting_layer += 1

    for z in range(starting_layer, starting_layer+block_height+1):
        # print('z')
        while z >= len(tower): tower.append(new_layer())
        for x in range(block[0]['x'], block[1]['x']+1):
            # print('x')
            for y in range(block[0]['y'], block[1]['y']+1):
                # print('y')
                tower[z][y][x] = str(blockI)
                # print('hi')
    
    return tower



def part1(data):
    data = [(parse_line(l)) for l in data]
    data = sorted(data, key= lambda x: x[0]['z']+0.1*x[1]['z'])
    print(data)
    groundLayer = [['#' for i in range(9)] for j in range(9)]
    tower = [groundLayer, new_layer()]

    for i, block in enumerate(data):
        tower = place_block(block, i, tower)
    
    removes = 0
    for i in range(len(tower[:-1])):
        if i == 0: continue # skip base layer
        layerA = tower[i]
        layerB = tower[i+1]
        setA = set(i for r in layerA for i in set(r))
        setB = set(i for r in layerB for i in set(r))

        lowerBlocks = (setA).difference(setB)
        if '.' in lowerBlocks: lowerBlocks.remove('.')
        if len(lowerBlocks) == 0: continue
        print('lowerblocks', lowerBlocks)
        

        for k in (setB).difference(setA): # for every block in layer B not in layer A
            supports = set()
            for x in range(10):
                for y in range(10):
                    if layerB[y][x] == k:
                        supports.add(layerA[y][x])
            if '.' in supports: supports.remove('.')
            if len(supports) == 1: 
                support = supports.pop()
                if support in lowerBlocks: lowerBlocks.remove(support)
        removes += len(lowerBlocks)
        print('retained:  ', lowerBlocks)

    print(set(i for r in tower[-1] for i in set(r)), len(set(i for r in tower[-1] for i in set(r))))
    removes += len(set(i for r in tower[-1] for i in set(r)))-1
    print(removes)


    

part1(puzzle_in)
# ! 1484 too high
```

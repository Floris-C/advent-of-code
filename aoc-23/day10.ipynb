```python
import queue
global data
data = example
data = [[c for c in line] for line in data]

def connections(x,y):
    symbol = data[y][x]
    match symbol:
        case '|': return [[x,y+1],[x,y-1]]
        case '-': return [[x+1,y],[x-1,y]]
        case 'L': return [[x+1,y],[x,y-1]]
        case 'J': return [[x-1,y],[x,y-1]]
        case '7': return [[x-1,y],[x,y+1]]
        case 'F': return [[x+1,y],[x,y+1]]
        case '.': return None
        case 'S': return None

def findEnclosedArea(x,y):
    unchecked = queue.SimpleQueue()
    unchecked.put([x,y])
    isEnclosed = True
    size = 0
    while not unchecked.empty():
        x,y = unchecked.get()
        if data[y][x] == 'X' or data[y][x] == 'O': continue
        data[y][x] = 'O'
        size += 1
        for d_x in range(-1,2):
            if (x + d_x < 0) or (x+d_x >= len(data[0])):
                isEnclosed = False
                continue
            for d_y in range(-1,2):
                if (y+d_y < 0) or (y+d_y >= len(data)):
                    isEnclosed = False
                    continue
                elif data[y+d_y][x+d_x] != 'X' and \
                    data[y+d_y][x+d_x] != 'O':
                    unchecked.put([x+d_x, y+d_y])
    return isEnclosed, size


# find starting position
for y, l in enumerate(data):
    if 'S' in l: 
        start = [l.index('S'), y]
        break
# find nodes connecting to S
x, y = start
next = queue.Queue()
for d_x in range(-1,2):
    for d_y in range(-1,2):
        conn = connections(x+d_x, y+d_y)
        if not conn: continue
        if start in conn:
            next.put_nowait([x+d_x, y+d_y])
data[y][x] = 'X'

maxSize = 0
while not next.empty():
    x,y = next.get_nowait()
    if data[y][x] == 'X': continue
    for c in connections(x,y):
        v = data[c[1]][c[0]]
        if v == 'X':
            data[y][x] = 'X'
        else: # if proper node
            next.put_nowait(c)

for d in data:
    print(d)

totalSize = 0
for y in range(len(data)):
    for x in range(len(data[y])):
        if data[y][x] != 'X' and data[y][x] != 'O': 
            isEnclosed, size = findEnclosedArea(x,y)
            if isEnclosed: totalSize += size

            print()
            for d in data:
                print(d)
            print(size, totalSize)

print()
for d in data:
    print(d)



print(totalSize)
```

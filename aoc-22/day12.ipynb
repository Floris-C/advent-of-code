```python
#day 12 AOC
inputFile = open("inputs/day12.txt", "r")
input = inputFile.read()
grid = [x for x in input.splitlines() if x != ""]
inputFile.close()
print(grid[:25])

alphabet = "abcdefghijklmnopqrstuvwxyz"

start, end = None, None
for i in range(len(grid)):
        for j in range(len(grid[i])): 
            if grid[i][j] == "S": start = {"x": j, "y": i, "val": grid[i][j]}
            if grid[i][j] == "E": end = {"x": j, "y": i, "val": grid[i][j]}
            
print(start, end)
```

```python
def accessible(curVal, targetVal):
    if curVal == "S": curVal = "a"
    if targetVal == "S": targetVal = "a"
    if curVal == "E": curVal = "z"
    if targetVal == "E": targetVal = "z"
    #print("access {} {}".format(curVal, targetVal))
    #print(alphabet)
    return (alphabet.index(curVal)+1) >= (alphabet.index(targetVal))



def getOptions(current): # cur = [x,y]
    x, y, val = current.values()
    options = []
    if x != 0:             options.append({"x": x-1, "y": y  , "val": grid[y][x-1]})     
    if x < len(grid[0])-1: options.append({"x": x+1, "y": y  , "val": grid[y][x+1]})
    if y != 0:             options.append({"x": x  , "y": y-1, "val": grid[y-1][x]})
    if y < len(grid)-1:    options.append({"x": x  , "y": y+1, "val": grid[y+1][x]})
    
    return options

def calcH(cur, end):
    return 0 # (cur["x"]-end["x"])**2 + (cur["y"]-end["y"])**2



def aStar():

    def addToOpen(node, f, openList):
        openList[0].append(node)
        openList[1].append(f)
        openList[2].append(f + calcH(node, end))
        return openList

    def findNextI(openList):
        minV = min(openList[2])
        return openList[2].index(minV)
    
    def updateOpen(i, option, f, end):
        openList[0][i] = option
        openList[0][i] = f
        openList[0][i] = f + calcH(option, end)
        return openList
    
    def moveToClosed(i, openList, closedList):
        closedList[0].append(openList[0].pop(i))
        closedList[1].append(openList[1].pop(i))
        closedList[2].append(openList[2].pop(i))
        return openList, closedList

    openListN = []
    openListF = []
    openListH = []
    openList = [openListN, openListF, openListH]
    closedList = [[],[],[]] # like open
    

    
    openList = addToOpen(start, 0, openList)
    #print(openList)

    while (not (end in closedList[0])) and (not (openList == [[],[],[]])):
        curI = findNextI(openList)
        cur, f, h = openList[0][curI], openList[1][curI], openList[2][curI]
        options = getOptions(cur)
        #print(options)
        #print("Options {}".format(options))
        for option in options:
            if accessible(cur["val"], option["val"]):
                #print(option)
                if not(option in openList[0] or option in closedList[0]):
                    openList = addToOpen(option, f+1, openList)
                elif option in openList[0]:
                    optionI = openList[0].index(option)
                    optionH = openList[2][optionI]
                    if optionH < h: openList =  updateOpen(i, option, f+1, openList)
            #else (not accessible) = do nothing
        openList, closedList = moveToClosed(curI, openList, closedList)
        #print(openList)

    return(closedList, end)

closed, target = aStar()
if target in closed[0]:
    i = closed[0].index(target)
    print("score: {}".format(closed[1][i]))
else:
    print("target wasn't reached")

#330 is too low
#394 too high
```

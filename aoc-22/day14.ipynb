```python
#day 12 AOC
inputFile = open("inputs/day14.txt", "r")
input = inputFile.read()
inputFile.close()



sand = "o"
air = "."
rock = "#"

maxY = 6
minY = 50
maxX = 450
minX = 500

rockPaths = []
for i in [x for x in input.splitlines() if x != ""]:
    currentPath = []
    for c in i.split(" -> "):
        x,y = c.split(",")
        x,y = int(x), int(y)

        currentPath.append((x, y))

        if x > maxX: maxX = x
        if x < minX: minX = x
        if y > maxY: maxY = y
        if y < minY: minY = y
    rockPaths.append(currentPath)

#print([x for x in input.splitlines() if x != ""][42])
print(rockPaths[0])
print(maxY, minY, maxX, minX)

```

```python
grid = [[air for x in range(1000)] for y in range(200)]

for path in rockPaths:
    for i in range(len(path)-1):
        #print(path[i])
        
        sx, sy = path[i]
        
        ex, ey = path[i+1]
        if ex < sx: sx, ex = ex, sx
        if ey < sy: sy, ey = ey, sy
        #print(sx, ex, sy, ey)
        for x in range(sx,ex+1):
            for y in range(sy, ey+1):#(x, y) in [(x,y) for x in range(sx,ex+1) for y in range(sy,ey+1)]:
                grid[y][x] = rock
                #print(x,y)

#print(grid)
keepgrid = grid.copy()
#print(grid[45][483])
```

```python
part2grid = keepgrid.copy()
floorY = maxY + 2
for i in range(len(grid[floorY])):
    part2grid[floorY][i] = rock
```

```python
grid = part2grid.copy()


sandSpawns = 0
endPoint = False

while not endPoint:
    sandX, sandY, falling = 500, 0, True
    sandSpawns += 1
    #print(sandSpawns)
    while falling:
        if grid[sandY+1][sandX] == air: #straight down
            sandX, sandY = sandX, sandY+1
        elif grid[sandY+1][sandX-1] == air: #diagonal left
            sandX, sandY = sandX-1, sandY+1
        elif grid[sandY+1][sandX+1] == air: #diagonal right
            sandX, sandY = sandX+1, sandY+1
        else: #stuck
            falling = False
    grid[sandY][sandX] = sand
    endPoint = sandY == 0 and sandX == 500

print(sandSpawns)
#26680 is TOO LOW XD
#27324 WAS CORRECT :D
for i in grid[0:floorY+1]:
    print("".join(i[minX-100:maxX+100]))
```

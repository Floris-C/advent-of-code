```python
# * Setup

monkeys = dict()
# monkey = 
# items = 
# operation = "+" int v "*" int
# divisibleTest = 
# nextFalse = 
# nextTrue = 
def doOperation(old, operation):
    if operation[1] == "old": 
        if operation[0] == "+":
            return old + old
        elif operation[0] == "*":
            return old * old
    if operation[0] == "+":
        return old + int(operation[1])
    elif operation[0] == "*":
        return old * int(operation[1])
    else: 
        print("ERROR: unkown operation") 
        return 0

def doTest(val, test):
    return val % test == 0

tests = []

for i in range(len(instructions)//7+1):
    monkey = int(instructions[i*7][7])
    monkeys[monkey] = dict()
    monkeys[monkey]["items"] = [int(x) for x in instructions[i*7+1][18:].split(", ")]
    monkeys[monkey]["operation"] = instructions[i*7+2][23:].split(" ")
    monkeys[monkey]["test"] = int(instructions[i*7+3].split("by ")[1])
    tests.append(int(instructions[i*7+3].split("by ")[1]))
    monkeys[monkey]["nextTrue"] = int(instructions[i*7+4].split("monkey ")[1])
    monkeys[monkey]["nextFalse"] = int(instructions[i*7+5].split("monkey ")[1])

testProd = 1
for t in tests:
    testProd *= t

for m in monkeys:
    print(m)
    print(monkeys[m])


```

```python
# Actually this is step 2 now

# The trick here is that the numbers increase without a limit which slows computation down
#  immensily and also will eventually cause integer overflow probably. Instead we need to find a way
# to make each items smaller while maintaining the divisibility of the items 


def monkeying(no, monkey):
    #print("MONKEY {}".format(no))
    #if no == 3: print("monkey 3 {}".format(monkey.get("items")))
    for item in monkey.get("items"):
        #print(item)
        business[no] += 1
        item = doOperation(item, monkey["operation"])
        item = item if item < testProd else item % testProd
        #item = item//3
        if doTest(item, monkey["test"]):
            #print("passed", no, item)
            monkeys[monkey["nextTrue"]]["items"].append(item)
        else:
            monkeys[monkey["nextFalse"]]["items"].append(item)
    monkey["items"] = []
    return monkey


business = [0 for i in range(len(monkeys))]
for round in range(10_000):
    for monkey in monkeys:
        monkeys[monkey] = monkeying(monkey, monkeys[monkey])
    if ((round+1))%100 == 0 or round == 0 or round == 19: 
        print("ROUND {}".format(round+1))
        print(business)
        for monkey in monkeys:
            print (monkeys[monkey])

    #print("round {}".format(round))
    #print(business)
    #for monkey in monkeys:
    #    print(monkeys[monkey]["items"])
    

print(business)
business.sort()
print(business)
print(business[-1]*business[-2])
```

```python
def evalLists(a,b):
    
    for i in range(len(a)):
        if i >= len(b):
            #print("right side ran out of items, false")
            return "invalid"
            
        if (type(a[i]) is int) and (type(b[i]) is int):
            if a[i] > b[i]: 
                #print("int int false")
                return "invalid"
            if a[i] < b[i]:
                #print("left smaller, so true")
                return "valid"
        
        elif (type(a[i]) is int) and (type(b[i]) is list):
            comparison = evalLists([a[i]], b[i])
            if comparison in ["valid", "invalid"]: return comparison
        
        elif (type(a[i]) is list) and (type(b[i]) is int):
            comparison = evalLists(a[i], [b[i]])
            if comparison in ["valid", "invalid"]: return comparison
        
        elif (type(a[i]) is list) and (type(b[i]) is list):
            comparison = evalLists(a[i], b[i])
            if comparison in ["valid", "invalid"]: return comparison
        else: print("FUCK")
    if (len(a) == len(b)): return "even"
    return "valid"
    print("made even (ie left side ran out of items)")
    
    return "even"


correctPairs = []
for index, [a,b] in enumerate(pairs):
    #print("--- pair {} ---".format(index+1))
    if evalLists(a,b) in ["valid", "even"]: correctPairs.append(index+1)
    #print()
    


sum(correctPairs)
#1233 is not right
#6473 is too high
#5997 is too low
#6272 was correct :D
```

```python
#day 9 AOC
inputFile = open("inputs/day9.txt", "r")
input = inputFile.read()
instructions = [] 
for i in [x for x in input.splitlines() if x != ""]:
    i = i.split(" ")
    newPair = (i[0], int(i[1]))
    instructions.append(newPair)
inputFile.close()
print(instructions[:5])
```

```python
#Part 1
# Let's say that the rope is placed on an infinite grid.
# We thus shouldn't model the grid as an image just retain a pair of positions for both the head and tail

# Point = (x, y)
head = [0, 0]
tail = [0, 0]

tailPositions = []
headPositions = []

def notNear(dif):
    return dif > 1 or dif < -1
def notNearDiagonal(xdif, ydif):
    return \
        (notNear(xDif) and yDif != 0) or \
        (notNear(yDif) and xDif != 0)

#We then go over the list of instructions.
eol = False
for instr in instructions:
    instr = list(instr)
    while instr[1] > 0:
        #move H
        match instr[0]:
            case "U": head[1] += 1
            case "R": head[0] += 1
            case "D": head[1] -= 1
            case "L": head[0] -= 1
        instr[1] -= 1
        headPositions.append(tuple(head))
        #print("H {}".format(head))
        #move T
        xDif = head[0]-tail[0]
        yDif = head[1]-tail[1]
        if notNearDiagonal(xDif, yDif):
            #print(head, tail, xDif, yDif)
            tail[0] += xDif // abs(xDif)
            tail[1] += yDif // abs(yDif)
        elif notNear(xDif):
            tail[0] += xDif // abs(xDif)
        elif notNear(yDif):
            tail[1] += yDif // abs(yDif)
        #print("T {}".format(tail))
        #save T position
        tailPositions.append(tuple(tail))
    #go next instr

#print(headPositions)
#print(tailPositions)
# print(list(set(tailPositions)))
print(len(list(set(tailPositions))))
# 11453 too high
```

```python
#day 4 AOC
inputFile = open("inputs/day4.txt", "r")
input = inputFile.read()
inputList = [] 
for i in [x for x in input.splitlines() if x != ""]:
    inputList.append(i)
inputFile.close()
print(inputList[:5])
```

```python
parsedList = []
for i in inputList:
    (a, b) = i.split(',')
    tempPair = (a.split('-'), b.split('-'))
    badPair = ([x for x in range(int(tempPair[0][0]), int(tempPair[0][1]))]
              ,[x for x in range(int(tempPair[1][0]), int(tempPair[1][1]))])
    parsedList.append(tempPair)
print(parsedList[:5])
```

```python
counter = 0
for i in range(len(parsedList)):
    for ii in range(len(parsedList[i])):
        for iii in range(len(parsedList[i][ii])):
            parsedList[i][ii][iii] = int(parsedList[i][ii][iii])
for(a, b) in parsedList:
    if   (a[0] >= b[0]) and (a[1] <= b[1]): 
        counter += 1
        print(a,b,1)
    elif (b[0] >= a[0]) and (b[1] <= a[1]): 
        counter += 1
        print(a,b,2)


print(counter)
print(len(parsedList))

#603 is too high --  contained ['11', '46'] ['8', '11']
```

```python
#day 8 AOC
import numpy as np
inputFile = open("inputs/day8.txt", "r")
input = inputFile.read()
inputStr =  [str(x) for x in input.splitlines() if x != ""]
inputFile.close()
print(inputStr[:10])
sizes = []
for i in inputStr:
    sizes.append(len(inputStr))
print(set(sizes))
print(len(inputStr))
```

```python
#Third attempt
visibility = []
for y in range(len(inputStr)):
    row = []
    for x in range(len(inputStr[y])):
        row.append(" ")
    visibility.append(row)

#rain from sides
for y in range(len(inputStr)):
    #from left
    visibility[y][0] = "x"
    max = inputStr[y][0]
    for x in range(1, len(inputStr[y])):
        if inputStr[y][x] > max: 
            max = inputStr[y][x]
            visibility[y][x] = "x"
    #from right
    visibility[y][len(inputStr[y])-1] = "x"
    max = inputStr[y][len(inputStr[y])-1]
    for x in reversed(range(0, len(inputStr[y])-1)):
        if inputStr[y][x] > max:  
            max = inputStr[y][x]
            visibility[y][x] = "x"

#vertical rain
for x in range(len(inputStr[0])):
    #from up
    visibility[0][x] = "x"
    max = inputStr[0][x]
    for y in range(1, len(inputStr)):
        if inputStr[y][x] > max: 
            max = inputStr[y][x]
            visibility[y][x] = "x"
    #from down
    visibility[len(inputStr[y])-1][x] = "x"
    max = inputStr[len(inputStr[y])-1][x]
    for y in reversed(range(0, len(inputStr)-1)):
        if inputStr[y][x] > max: 
            max = inputStr[y][x]
            visibility[y][x] = "x"


#display 3
counter = 0
for y in visibility:
    for x in y:
        if x == "x": counter += 1
        print(x, end = "")
    print("")
print("No1: {}".format(counter))

# xxxxx
# xxx x
# xx xx
# x x x
# xxxxx
```

```python
def scenic_score(x,y):
    height = inputStr[y][x]
    
    right = 0 
    nx = x+1
    while nx < len(inputStr[y]):
        right += 1
        nx = nx + 1 if inputStr[y][nx] < height else len(inputStr)
    
    left = 0
    nx = x-1
    while nx >= 0:
        left += 1
        nx = nx - 1 if inputStr[y][nx] < height else -1

    down = 0 
    ny = y+1
    while ny < len(inputStr[x]):
        down += 1
        ny = ny + 1 if inputStr[ny][x] < height else len(inputStr[x])
    
    up = 0
    ny = y-1
    while ny >= 0:
        up += 1
        ny = ny - 1 if inputStr[ny][x] < height else -1
        

        

    # print(left, right, up, down)
    return left*right*up*down


max = 0
for y in range(len(inputStr)):
    for x in range(len(inputStr[y])):
        score = scenic_score(x,y)
        if score > max: max = score
        # print(score, end = " ")
print("No2: {}".format(max))
```

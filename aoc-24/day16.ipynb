```python
def possible_moves(x, y, h):
    match h:
        case 'N':
            return [(x, y, 'E'), (x, y, 'W'), (x, y-1, h)]
        case 'E':
            return [(x, y, 'N'), (x, y, 'S'), (x+1, y, h)]
        case 'S':
            return [(x, y, 'E'), (x, y, 'W'), (x, y+1, h)]
        case 'W':
            return [(x, y, 'N'), (x, y, 'S'), (x-1, y, h)]
        case _:
            raise Exception("wtf no heading for possible moves")

def task1(input):
    grid = [[c if c == '#' else {} for c in line] for line in input]
    sx, sy = 1, len(grid)-2
    ex, ey = len(grid[1])-2, 1

    queue = [(sx, sy, 'E', 0)]

    while queue:
        x, y, h, c = queue.pop(0)
        #if the move goes into a wall, don't bother
        if grid[y][x] == '#': continue
        #if the move is more expensive than an extensive alternative, don't bother
        if h in grid[y][x] and c > grid[y][x][h]:
            continue
        #else register the move and keep going
        grid[y][x][h] = c
        for move in possible_moves(x,y,h):
            nc = c+1 if move[2] == h else c+1000
            queue.append((*move, nc))
    
    return grid, min(grid[ey][ex].values())

task1_grid, ans = task1(input)
print(ans)
```

```python
def possible_moves_reverse(x, y, h):
    match h:
        case 'N':
            return [(x, y, 'E'), (x, y, 'W'), (x, y+1, h)]
        case 'E':
            return [(x, y, 'N'), (x, y, 'S'), (x-1, y, h)]
        case 'S':
            return [(x, y, 'E'), (x, y, 'W'), (x, y-1, h)]
        case 'W':
            return [(x, y, 'N'), (x, y, 'S'), (x+1, y, h)]
        case _:
            raise Exception("wtf no heading for possible moves")

def task2(grid):
    # grid, _ = task1(example)
    sx, sy = 1, len(grid)-2
    ex, ey = len(grid[1])-2, 1
    
    new_grid = [['#' if c == '#' else '.' for c in line] for line in grid]
    new_grid[ey][ex] = 'O'
    h, c = min(grid[ey][ex].items(), key=lambda x: x[1])
    queue = [(ex, ey, h, c)]

    while queue:
        x, y, h, c = queue.pop(0)
        for nx, ny, nh in possible_moves_reverse(x,y,h):
            if grid[ny][nx] == '#': continue
            nc = grid[ny][nx][nh]
            if (nh != h and c==nc+1000) or (h == nh and c==nc+1):
                queue.append((nx, ny, nh, nc))
                new_grid[ny][nx] = 'O'

    return sum([sum([c == 'O' for c in line]) for line in new_grid])


task2(task1_grid)
```

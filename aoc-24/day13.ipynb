```python
class ClawMachine():
    def __init__(self, ax, ay, bx, by, rx, ry):
        self.a_weight = 3
        self.b_weight = 1
        self.ax = ax
        self.ay = ay
        self.bx = bx
        self.by = by
        self.rx = rx
        self.ry = ry

    def solve(self):
        ap = (self.rx - (self.bx / self.by) * self.ry) / (self.ax - self.bx * self.ay / self.by)
        bp = (self.ry - self.ay * ap) / self.by
        return round(ap), round(bp), ap, bp
    
    def check_solve(self, ap, bp):
        return  ap*self.ax + bp*self.bx == self.rx and \
                ap*self.ay + bp*self.by == self.ry


def task1(input):
    tokens_spent = 0
    for i in range(0, len(input), 4):
        ax, ay = map(int, (lambda e: (e[1].split(',')[0], e[2]))(input[i].split('+')))
        bx, by = map(int, (lambda e: (e[1].split(',')[0], e[2]))(input[i+1].split('+')))
        rx, ry = map(int, (lambda e: (e[1].split(',')[0], e[2]))(input[i+2].split('=')))
        claw_machine = ClawMachine(ax, ay, bx, by, rx, ry)
        ap, bp, apf, bpf = claw_machine.solve()

        if ap <= 100 and bp <= 100 and claw_machine.check_solve(ap, bp):
            tokens_spent += 3*ap + bp
        print(ax, ay, bx, by, rx, ry, ap, bp, tokens_spent)
        print(apf, bpf, claw_machine.check_solve(ap, bp))
    return tokens_spent
        

task1(input)
#22974 is too low...
#25260 is STILL too low...
```
